```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

def gaussian_kernel_1d(size: int, sigma: float) -> np.ndarray:
    """Generate a 1D Gaussian kernel."""
    x = np.linspace(-size // 2, size // 2, size)
    kernel = np.exp(-x**2 / (2 * sigma**2))
    return kernel / kernel.sum()

def piecewise_parabolic_kernel_1d(size: int, sigma: float) -> np.ndarray:
    """Generate a 1D piecewise parabolic kernel with compact support."""
    x = np.linspace(-sigma, sigma, size)
    
    def f(x):
        """Piecewise function f(x)"""
        if -sigma <= x < -sigma / 2:
            return (2 / sigma) * (1 + x / sigma)**2
        elif -sigma / 2 <= x <= sigma / 2:
            return (1 / sigma) - (2 / sigma) * (x / sigma)**2
        elif sigma / 2 < x <= sigma:
            return (2 / sigma) * (1 - x / sigma)**2
        else:
            return 0

    kernel = np.array([f(xi) for xi in x])
    return kernel / kernel.sum()  # Normalize

def double_well_potential(x: np.ndarray) -> np.ndarray:
    """Compute the 1D double-well potential function U(x) = x^4 - 4x^2 + 0.2x."""
    return x**4 - 4*x**2 + 0.2*x

# Parameters
size = 101  # Number of points in the kernel
sigma = 2.0  # Width of the kernel

# Generate the 1D Gaussian kernel
gaussian_kernel = gaussian_kernel_1d(size, sigma)

# Generate the 1D piecewise parabolic kernel
pp_kernel = piecewise_parabolic_kernel_1d(size, sigma)

# Define the range and sample points
x = np.linspace(-2, 2, 500)

# Compute the double-well potential
potential_array = double_well_potential(x)

# Perform convolutions
conv_gaussian = convolve(potential_array, gaussian_kernel, mode='same')
conv_pp = convolve(potential_array, pp_kernel, mode='same')

# Plotting the results
plt.figure(figsize=(16, 8), facecolor='white')

# Original potential
plt.subplot(1, 3, 1)
plt.plot(x, potential_array, label='Double-Well Potential', color='blue')
plt.title('Original Double-Well Potential')
plt.xlabel('x')
plt.ylabel('Free Energy (kBT)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.gca().set_facecolor('white')

# Gaussian convolution
plt.subplot(1, 3, 2)
plt.plot(x, conv_gaussian, label='Gaussian Convolution', color='red')
plt.plot(x, conv_gaussian, 'o', label='Gaussian Convolution', color='red', markevery=20)
plt.title('Gaussian Convolution')
plt.xlabel('x')
plt.ylabel('Free Energy (kBT)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.gca().set_facecolor('white')

# Piecewise parabolic convolution
plt.subplot(1, 3, 3)
plt.plot(x, conv_pp, label='Piecewise Parabolic Convolution', color='green')
plt.plot(x, conv_pp, 's', label='Piecewise Parabolic Convolution', color='green', markevery=20)
plt.title('Piecewise Parabolic Convolution')
plt.xlabel('x')
plt.ylabel('Free Energy (kBT)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.gca().set_facecolor('white')

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('double_well_potential_convolution_subplot_white_bg.png', facecolor='white')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

def gaussian_kernel_1d(size: int, sigma: float) -> np.ndarray:
    """Generate a 1D Gaussian kernel."""
    x = np.linspace(-size // 2, size // 2, size)
    kernel = np.exp(-x**2 / (2 * sigma**2))
    return kernel / kernel.sum()

def piecewise_parabolic_kernel_1d(size: int, sigma: float) -> np.ndarray:
    """Generate a 1D piecewise parabolic kernel with compact support."""
    x = np.linspace(-sigma, sigma, size)
    
    def f(x):
        """Piecewise function f(x)"""
        if -sigma <= x < -sigma / 2:
            return (2 / sigma) * (1 + x / sigma)**2
        elif -sigma / 2 <= x <= sigma / 2:
            return (1 / sigma) - (2 / sigma) * (x / sigma)**2
        elif sigma / 2 < x <= sigma:
            return (2 / sigma) * (1 - x / sigma)**2
        else:
            return 0

    kernel = np.array([f(xi) for xi in x])
    return kernel / kernel.sum()  # Normalize

def double_well_potential(x: np.ndarray) -> np.ndarray:
    """Compute the 1D double-well potential function U(x) = x^4 - 4x^2 + 0.2x."""
    return x**4 - 4*x**2 + 0.2*x

def equilibrium_density(potential: np.ndarray) -> np.ndarray:
    """Compute the equilibrium density from the potential."""
    density = np.exp(-potential)
    return density / density.sum()

# Parameters
size = 101  # Number of points in the kernel
sigmas = [0.2, 0.3, 0.4]  # Different sigma values

# Define the range and sample points
x = np.linspace(-2, 2, 500)

# Compute the double-well potential
potential_array = double_well_potential(x)

# Compute the equilibrium density
equilibrium_density_array = equilibrium_density(potential_array)

# Plotting the results
plt.figure(figsize=(16, 12), facecolor='white')

for i, sigma in enumerate(sigmas):
    # Generate the 1D Gaussian kernel
    gaussian_kernel = gaussian_kernel_1d(size, sigma)
    
    # Generate the 1D piecewise parabolic kernel
    pp_kernel = piecewise_parabolic_kernel_1d(size, sigma)
    
    # Perform convolutions
    conv_gaussian = convolve(equilibrium_density_array, gaussian_kernel, mode='same')
    conv_pp = convolve(equilibrium_density_array, pp_kernel, mode='same')
    
    # Plotting the densities
    plt.subplot(3, 1, i + 1)
    plt.plot(x, equilibrium_density_array, label='Equilibrium Density', color='blue', alpha=0.5)
    plt.plot(x, conv_gaussian, label=f'Gaussian Conv. ($\sigma={sigma}$)', color='red')
    plt.plot(x, conv_gaussian, 'o', color='red', markevery=20)
    plt.plot(x, conv_pp, label=f'PP Kernel Conv. ($\sigma={sigma}$)', color='green')
    plt.plot(x, conv_pp, 's', color='green', markevery=20)
    plt.title(f'Densities Convolution ($\sigma={sigma}$)')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.gca().set_facecolor('white')

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('double_well_density_convolution_effect.png', facecolor='white')
plt.show()
```
