```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

def gaussian_kernel_2d(size: int, sigma: float) -> np.ndarray:
    """Generate a 2D Gaussian kernel."""
    x = np.linspace(-size // 2, size // 2, size)
    X, Y = np.meshgrid(x, x)
    kernel = np.exp(-(X**2 + Y**2) / (2 * sigma**2))
    return kernel / kernel.sum()

def piecewise_parabolic_kernel_2d(size: int, sigma: float) -> np.ndarray:
    """Generate a 2D piecewise parabolic kernel with compact support."""
    x = np.linspace(-sigma, sigma, size)
    X, Y = np.meshgrid(x, x)

    def f(xi, yi):
        """Piecewise function f(xi, yi)"""
        if -sigma <= xi < -sigma / 2 and -sigma <= yi < -sigma / 2:
            return (2 / sigma**2) * (1 + xi / sigma)**2 * (1 + yi / sigma)**2
        elif -sigma / 2 <= xi <= sigma / 2 and -sigma / 2 <= yi <= sigma / 2:
            return (1 / sigma**2) - (2 / sigma**2) * (xi / sigma)**2 - (2 / sigma**2) * (yi / sigma)**2
        elif sigma / 2 < xi <= sigma and sigma / 2 < yi <= sigma:
            return (2 / sigma**2) * (1 - xi / sigma)**2 * (1 - yi / sigma)**2
        else:
            return 0

    kernel = np.array([[f(xi, yi) for yi in x] for xi in x])
    return kernel / kernel.sum()

def entropic_double_well_potential_2d(x: np.ndarray, y: np.ndarray, sigma_x: float = 0.1, sigma_y: float = 0.1, a: float = 1.0) -> np.ndarray:
    """Compute the 2D Entropic Double-Well Potential."""
    return a * (x**6 + y**6 + np.exp(-((x / sigma_x) ** 2)) * (1 - np.exp(-((y / sigma_y) ** 2))))

def equilibrium_density_2d(potential: np.ndarray) -> np.ndarray:
    """Compute the equilibrium density from the 2D potential."""
    density = np.exp(-potential)
    return density / density.sum()

# Parameters
size = 51  # Kernel size
sigmas = [0.1, 0.2, 0.3]  # Different sigma values

# Define the range and sample points
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)

# Compute the Entropic Double-Well Potential
potential_array = entropic_double_well_potential_2d(X, Y)

# Compute the equilibrium density
equilibrium_density_array = equilibrium_density_2d(potential_array)

# Compute convolutions
conv_gaussian_dict = {}
conv_pp_dict = {}

for sigma in sigmas:
    # Generate the 2D Gaussian and piecewise parabolic kernels
    gaussian_kernel = gaussian_kernel_2d(size, sigma)
    pp_kernel = piecewise_parabolic_kernel_2d(size, sigma)
    
    # Perform convolutions
    conv_gaussian = convolve2d(equilibrium_density_array, gaussian_kernel, mode='same', boundary='wrap')
    conv_pp = convolve2d(equilibrium_density_array, pp_kernel, mode='same', boundary='wrap')
    
    # Store convolutions for each sigma
    conv_gaussian_dict[sigma] = conv_gaussian
    conv_pp_dict[sigma] = conv_pp

# Determine the color scale for consistent visualization
vmin = min(np.min(equilibrium_density_array), 
           *[np.min(conv) for conv in conv_gaussian_dict.values()],
           *[np.min(conv) for conv in conv_pp_dict.values()])
vmax = max(np.max(equilibrium_density_array), 
           *[np.max(conv) for conv in conv_gaussian_dict.values()],
           *[np.max(conv) for conv in conv_pp_dict.values()])

# Create a figure with a grid layout
fig, axes = plt.subplots(3, 3, figsize=(18, 18), gridspec_kw={'width_ratios': [1, 1, 1], 'wspace': 0.1}, constrained_layout=False, facecolor='white')

# Add a single colorbar
cax = fig.add_axes([0.92, 0.1, 0.02, 0.8])  # [x, y, width, height]

for i, sigma in enumerate(sigmas):
    # Plot equilibrium density
    im1 = axes[i, 0].imshow(equilibrium_density_array, extent=[-1.5, 1.5, -1.5, 1.5], origin='lower', cmap='viridis', vmin=vmin, vmax=vmax)
    axes[i, 0].set_title(f'Equilibrium Density ($\sigma={sigma}$)')
    axes[i, 0].set_xlabel('X')
    axes[i, 0].set_ylabel('Y')
    
    # Plot Gaussian convolution
    im2 = axes[i, 1].imshow(conv_gaussian_dict[sigma], extent=[-1.5, 1.5, -1.5, 1.5], origin='lower', cmap='viridis', vmin=vmin, vmax=vmax)
    axes[i, 1].set_title(f'Gaussian Conv. ($\sigma={sigma}$)')
    axes[i, 1].set_xlabel('X')
    axes[i, 1].set_ylabel('Y')
    
    # Plot Piecewise Parabolic convolution
    im3 = axes[i, 2].imshow(conv_pp_dict[sigma], extent=[-1.5, 1.5, -1.5, 1.5], origin='lower', cmap='viridis', vmin=vmin, vmax=vmax)
    axes[i, 2].set_title(f'Piecewise Parabolic Conv. ($\sigma={sigma}$)')
    axes[i, 2].set_xlabel('X')
    axes[i, 2].set_ylabel('Y')

# Add colorbar
cbar = fig.colorbar(im3, cax=cax)
cbar.set_label('Density')

# Save the figure
plt.savefig('entropic_double_well_density_convolutions_densities.png', facecolor='white')
plt.show()
```
