```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Define the potential function V(x)
def potential(x):
    return x**4 - 4*x**2 + 0.2*x

# Calculate the equilibrium density π(x)
def equilibrium_density(x, beta=1.0):
    V_x = potential(x)
    Z = np.trapz(np.exp(-beta * V_x), x)  # Normalization constant
    return np.exp(-beta * V_x) / Z

# Parameters
beta = 1.0
x = np.linspace(-2, 2, 1000)

# Calculate equilibrium density
pi_x = equilibrium_density(x, beta)

# Create Gaussian kernel for convolution
def gaussian_kernel(size, sigma):
    kernel = np.linspace(-size, size, 2*size+1)
    kernel = np.exp(-kernel**2 / (2 * sigma**2))
    kernel /= np.sum(kernel)
    return kernel

# Parameters for Gaussian convolution
sigma = 10
kernel_size = int(3 * sigma)  # Typical choice for the kernel size
gaussian_kernel = gaussian_kernel(kernel_size, sigma)

# Perform Gaussian convolution to smooth the density
smoothed_pi_x = convolve(pi_x, gaussian_kernel, mode='same')

# Plot the potential, equilibrium density, and smoothed density
plt.figure(figsize=(10, 6))

# Plot potential
plt.plot(x, potential(x), label='Potential V(x)', color='blue')

# Plot equilibrium density
plt.plot(x, pi_x, label='Equilibrium Density π(x)', color='green')

# Plot smoothed equilibrium density
plt.plot(x, smoothed_pi_x, label='Smoothed Density', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Potential, Equilibrium Density, and Smoothed Density')
plt.xlim(-2, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Define the potential function V(x)
def potential(x):
    return x**4 - 4*x**2 + 0.2*x

# Calculate the equilibrium density π(x)
def equilibrium_density(x, beta=1.0):
    V_x = potential(x)
    Z = np.trapz(np.exp(-beta * V_x), x)  # Normalization constant
    return np.exp(-beta * V_x) / Z

# Create Gaussian kernel for convolution
def gaussian_kernel(size, sigma):
    kernel = np.linspace(-size, size, 2*size+1)
    kernel = np.exp(-kernel**2 / (2 * sigma**2))
    kernel /= np.sum(kernel)
    return kernel

# Parameters
beta = 1.0
x = np.linspace(-2, 2, 1000)

# Calculate equilibrium density
pi_x = equilibrium_density(x, beta)

# List of bandwidths (sigma values)
sigma_values = [0.2, 0.4, 0.6, 1.0]

# Create a single subplot for smoothed densities
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the equilibrium density
ax.plot(x, pi_x, label='Equilibrium Density π(x)', color='green', linestyle='--')

# Plot smoothed densities for different bandwidths
for sigma in sigma_values:
    kernel_size = int(5 * sigma)  # Adjust kernel size based on sigma
    kernel = gaussian_kernel(kernel_size, sigma)
    smoothed_pi_x = convolve(pi_x, kernel, mode='same')
    ax.plot(x, smoothed_pi_x, label=f'Smoothed Density (σ={sigma})')

# Configure plot
ax.set_xlabel('x')
ax.set_ylabel('Density')
ax.set_title('Equilibrium and Smoothed Densities for Different Bandwidths')
ax.legend()
ax.grid(True)
ax.set_xlim(-2, 2)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig('density_comparison_bandwidths.png', dpi=300)  # Save as PNG with 300 dpi

# Display the figure
plt.show()
```
